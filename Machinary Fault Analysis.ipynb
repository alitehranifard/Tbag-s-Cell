## Handling Errors and Outliers

```python
means = cleaned_data.mean()
print(means)
```

```python
stds = cleaned_data.std()
print(stds)
```

```python
threshold = 3
outliers = cleaned_data[(np.abs((cleaned_data - means) / stds) > threshold).any(axis=1)]
print('\nIdentified Outliers:')
print('\n',outliers)
data_no_outliers = cleaned_data[(np.abs((cleaned_data - means) / stds) <= threshold).all(axis=1)]
print('\nDataset after Removing Outliers\n',data_no_outliers)
```

```python
df = data_no_outliers 
print('\n',df.shape)
```

```python
df.head(10)
```

```python
df['rolling_mean_A'] = df['A'].rolling(window=20, min_periods=5).mean()
df['rolling_std_A'] = df['A'].rolling(window=20, min_periods=5).std()
df['rolling_mean_B'] = df['B'].rolling(window=20, min_periods=5).mean()
df['rolling_std_B'] = df['B'].rolling(window=20, min_periods=5).std()
df['rolling_mean_C'] = df['C'].rolling(window=20, min_periods=5).mean()
df['rolling_std_C'] = df['C'].rolling(window=20, min_periods=5).std()
df['rolling_mean_D'] = df['D'].rolling(window=20, min_periods=5).mean()
df['rolling_std_D'] = df['D'].rolling(window=20, min_periods=5).std()
df['rolling_mean_E'] = df['E'].rolling(window=20, min_periods=5).mean()
df['rolling_std_E'] = df['E'].rolling(window=20, min_periods=5).std()
df['rolling_mean_F'] = df['F'].rolling(window=20, min_periods=5).mean()
df['rolling_std_F'] = df['F'].rolling(window=20, min_periods=5).std()
df['rolling_mean_G'] = df['G'].rolling(window=20, min_periods=5).mean()
df['rolling_std_G'] = df['G'].rolling(window=20, min_periods=5).std()
df['rolling_mean_H'] = df['H'].rolling(window=20, min_periods=5).mean()
df['rolling_std_H'] = df['H'].rolling(window=20, min_periods=5).std()
```

## implenting ANN 

```python
import tensorflow as tf
```

```python
X = shuffled_data.iloc[:, :-1].values
y = shuffled_data.iloc[:, -1].values
```

```python
print(y)
```

```python
from sklearn.preprocessing import OneHotEncoder
y = np.array(y).reshape(-1, 1)
encoder = OneHotEncoder(sparse_output=False)
y = encoder.fit_transform(y)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
print(X_train.shape)
```

```python
print(y_train.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
ann = tf.keras.models.Sequential()
ann.add(tf.keras.layers.Dense(units=4, activation='relu'))
ann.add(tf.keras.layers.Dense(units=4, activation='relu'))
ann.add(tf.keras.layers.Dense(units=3, activation='softmax'))
ann.compile(optimizer = 'adam' , loss = 'categorical_crossentropy' , metrics = ['accuracy'])
from keras.callbacks import EarlyStopping

early_stopping = EarlyStopping(
    monitor='accuracy',  # or 'val_accuracy' depending on your focus
    patience=2,         # number of epochs with no improvement to wait before stopping
    restore_best_weights=True  # restore weights from the best epoch
)
```

```python
from keras.callbacks import EarlyStopping

early_stopping = EarlyStopping(
    monitor='accuracy',  # or 'val_accuracy' depending on your focus
    patience=2,         # number of epochs with no improvement to wait before stopping
    restore_best_weights=True  # restore weights from the best epoch
)
```

```python
ann.fit(X_train, y_train, batch_size = 128, epochs = 20, callbacks=[early_stopping])
```

```python
y_pred = ann.predict(X_test)
y_pred = (y_pred > 0.5)
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
